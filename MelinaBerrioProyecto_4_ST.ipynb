### **K: exponential increase & comparison with real model** 

```python
def calculate_k(lista_paises):
    lista_k = [] #best values of k for every country
    lista_t0 = []
    lista_population = []

    #plot style
    fig, axes = plt.subplots(2, figsize=(15, 12))
    fig.suptitle('Covid19 Cases', fontsize=20)
    fig.tight_layout(pad=4.0)

    axes[0].set_title('Logaritmic behaviour',fontsize=14, loc='center')
    axes[0].set_xlim([0, 65])
    axes[0].grid()

    axes[1].set_title('Simulation',fontsize=14, loc='center')
    axes[1].set_xlim([0, 65])
    axes[1].grid()
    
    marker = itertools.cycle(('o', 'v', '^', '<', '>', 's', '8', 'p'))

    for x_pais in lista_paises:
        df = data2[(data2["location"].isin([x_pais]))&(pd.notnull(data2['total_cases']))]
        df=df.head(60)
        days=np.arange(0,df.shape[0])
        star_date=(df.iloc[:1]).date.tolist()[0]
        print("In {} The first case was registerd in {}".format(x_pais,star_date))
        color = next(axes[1]._get_lines.prop_cycler)['color']
        
        poblacion_xpais=df.population.tolist()[0]
    
        days=np.arange(0,df.shape[0])
        axes[0].plot(days,np.log(df.total_cases),label=x_pais, color=color)
        axes[0].set_xlabel('Days')
        axes[0].set_ylabel('log Total Cases')
        axes[0].legend()
    
        ################# K stimation
        popt, _ = curve_fit(exponencial, days, df.total_cases)
        lista_k.append(popt[0])
        lista_t0.append(popt[1])
        
        lista_population.append(poblacion_xpais)

        k=popt[0]
        t0=popt[1]
        casos_simu=exponencial(days,k,t0)
        
        axes[1].plot(days,df.total_cases,'.', color=color)
        axes[1].plot(days,casos_simu,label=x_pais, color = color)
        axes[1].set_xlabel('Days')
        axes[1].set_ylabel('Number of confirmed cases')
        axes[1].legend()

    return lista_k,lista_t0,lista_population
```

### Resampling: Bootstrap

```python
len(lista_k)
```

```python
def bootstraping(x, n_iteraciones=9999):   
    #n = len(x)  #/4 subset subset from population
    n=4
    dist_boot = np.full(shape=n_iteraciones, fill_value=np.nan)    
    for i in tqdm(range(n_iteraciones)):
        resample = np.random.choice(x, size=n, replace=True)  
        dist_boot[i] = np.mean(resample)  #stat
    return dist_boot
```

```python
ks = np.array(lista_k)
dist_boot = bootstraping(x = ks, n_iteraciones   = 9999)
```

```python
# IC con alpha=0.05 then 95% should go from quantil 0.025 to 0.975, accepting error of 5%
#IC for k representing worldwide population
cuantiles = np.quantile(a = dist_boot, q = [0.025, 0.975])
media_obs = ks.mean()
media_obs = "{:.3f}".format(ks.mean())
media_obs = "Mean: " + media_obs

int_obs_izq = list(cuantiles)
int_obs_izq = "{:.3f}".format(cuantiles[0])
int_obs_der = "{:.3f}".format(cuantiles[1])

IC_95 = "IC_95%: " + int_obs_izq +" a "+ int_obs_der

fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(7,4))
ax.hist(dist_boot, bins=30, density=True, color='#3182bd', alpha=0.5)
ax.axvline(x=ks.mean(), color='firebrick', label=media_obs)
ax.axvline(x=cuantiles[0], color='black', linestyle='--', label=IC_95)
ax.axvline(x=cuantiles[1], color='black', linestyle='--')
ax.hlines(y=0.001, xmin=cuantiles[0], xmax=cuantiles[1], color='black')
ax.set_title(' IC for k (dist bootstrapping)')
ax.set_xlabel('Mean')
ax.set_ylabel('Density')
ax.legend();
```

## Evaluating Strategies

**Policy responses chosen: Quarantine**

In `Policy responses` with variable `stringency_index`: Government Response Stringency Index: composite measure based on 9 response indicators including school closures, workplace closures, and travel bans, rescaled to a value from 0 to 100 (100 = strictest response)

**Indicators chosen:**
- paramteres of exponential growth:k and t0, since they explain the positive total cases of each country
- Reproduction rate: Rt is the average number of people infected by a single infected individual at a particular time t, by definition should has some relation with positive cases. the indicator will be the max of the reproduction reat founded in initial covid stage.
- new deaths, could be a influence factor for a country to decide to apply a high quarantine.
- number of days without quarantine, since a country with bigger initial period with no quarantine could have been forced to apply a severe quarantine if new deaths had increase exponentially.

```python
from math import isnan

# Viene limitado a paises donde exista el parametro reproduction rate 
# se hayan contabilizado new deaths

def review_country(x_pais):
   
    df = data1[(data1["location"].isin([x_pais]))]
    df=df.sort_values(by='date')
    first_index=int(df.index[0])
    first_index_notnan=df[["new_deaths"]].apply(pd.Series.first_valid_index).min()
    diff=first_index_notnan-first_index
    deaths=df.new_deaths
    df=df[(pd.notnull(df['total_cases']))]
    df=df.head(60)
    df.reset_index(inplace=True)
    rep=df.reproduction_rate
    casos_xpais=df.total_cases
    if (np.log(deaths.max())>0) and (not all(rep.isnull())) and (diff):
        lst=casos_xpais.tolist()
        days=np.arange(0,df.shape[0])
        popt, _ = curve_fit(exponencial, days, lst)
        k=popt[0]
        t0=popt[1]
        if t0 > 0:
            t0 = math.log(t0)
        elif t0 < 0:
            t0 = math.log(t0*-1)*-1
        elif t0 == 0:
            t0 = 0        
        var_k=variabilidad(deaths.mean(),diff)
        rep = rep.mean()
        deaths = np.log(deaths.mean())
        diff=np.log(diff)
        quarantine=df_order[df_order.location==x_pais].quarantine.values
        df_analizar_pais = pd.DataFrame({   'tasa_rep': rep,
                                            'k': k,
                                            't0': t0,
                                            'deaths': deaths,
                                            'num_days':diff,
                                            'quarantine': quarantine},index=[x_pais])
        return df_analizar_pais
    else:
        return 0  
```

```python
def sigmoid_function(x,w0, w1):
    z = w0 + w1*x 
    y = 1/(1 + np.exp(-z))   
    return y

def logistic_regression(indicators,traindf,testdf,countries_test):
    accuracy_train=[]
    predictions_train=[]
    accuracy_test=[]
    predictions_test=[]
    
    for i in indicators:
        fig, axes = plt.subplots(1,2, figsize=(25, 8))
        
        #training
        X1 = traindf[i].values.reshape(-1,1)
        y1 = traindf['quarantine'].values.reshape(-1,1)
        
        regresion_logistica1 = LogisticRegression()
        regresion_logistica1.fit(X1,y1)
        prediction = regresion_logistica1.predict(X1)
        predictions_train.append(prediction)
        accuracy = accuracy_score(y1, prediction)
        accuracy_train.append(accuracy)
        parameters=[regresion_logistica1.coef_, regresion_logistica1.intercept_]
        
        #graph
        X = traindf[i].values
        y = traindf['quarantine'].values
        label="Quarantine Vs "+i
        axes[0].set_title(label)
        axes[0].scatter(X, y, c = y, alpha = 0.5, lw = 0.5, edgecolor = 'red', cmap =cm.coolwarm)
        x_plot = X1
        axes[0].plot(x_plot, sigmoid_function(x_plot,regresion_logistica1.intercept_[0], regresion_logistica1.coef_[0][0]))
        axes[0].set_ylabel("Quarantine")
        axes[0].set_xlabel(i)
        
        # #probabilities
        probabilidades_prediccion2 = regresion_logistica1.predict_proba(X1)
        label2="Prediction Probabilities Vs "+i
        axes[1].set_title(label2)
        x_plot = X1
        y_plot = probabilidades_prediccion2[:,1]
        axes[1].scatter(x_plot, y_plot, alpha = 0.5, color = 'red')
        axes[1].set_xlabel(i)
        axes[1].set_ylabel('Prediction Probabilities')
        plt.show()        
        
        #testing
        X_test = testdf[i].values.reshape(-1,1)
        y_test = testdf['quarantine'].values.reshape(-1,1)
        prediction1 = regresion_logistica1.predict(X_test)
        predictions_test.append(prediction1)
        accuracy1 = accuracy_score(y_test, prediction1)
        accuracy_test.append(accuracy1)
    return accuracy_train,predictions_train,accuracy_test,predictions_test
```

### Final Inform

On first section the goal was to understand the curve of spread disease in a worldwide level, selecting a group of countries of hemisphere north (where the pandemic has started) and control if the spread behavior had been replicated in the rest of the world.
The logic behind:
$C(t)=e^{k(t-t_0)}$

Where: $t_0$ date of first positive case, y $k$ exponential factor of disease is directly proportional to the number of positive cases. Also depends on the level of contagiousness of the virus, that's why it can be evaluated with the quarantine level achieved.

It has been applied a bootstrapping technique to generalize the k parameter, to obtain a confidence interval and proceed to evaluate other populations, with results around k mean.
Also, it has been done a general evaluation to worldwide population, proving that in both cases exist a relation between k and the quantity of total cases.

It has been evaluated the level of quarantine for each country in their first 60 days with positive cases, having in consideration que quantity of days without any policy response; also, the mean of reproduction rate, and the number of new deaths.
With these indicators it had been constructed the logistic regression model for every indicator, and then trying initially for all parameters and by slices of them resulting in only 3: k parameter, reproduction rate, and number of new deaths. Being able to predict if a country has done a high quarantine or not. 
The data used for this classifier has considered only countries with useful data meaning that at least some observations should have reproduction rate, from the period before to registered confirmed positive cases I have choose the new deaths indicator, to prove if this have influence in the government decision to apply a high quarantine. By the results it has been confirmed.
